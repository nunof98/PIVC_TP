### Imports

```python
import os
import cv2
import pytesseract
from matplotlib import pyplot as plt
```

### Tesseract path

```python
tesseract_path = "C:/Program Files/Tesseract-OCR/tesseract.exe"
pytesseract.pytesseract.tesseract_cmd = tesseract_path
```

### Read image

```python
folder = "dataset"
img = cv2.imread(os.path.join(folder, "ui_000.png"))
plt.imshow(img)
```

### Convert to grayscale

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap='gray')
```

### Convert to binary trough fixed threshold

```python
_, img_tresh = cv2.threshold(img_gray, 10, 255, cv2.THRESH_BINARY | cv2.THRESH_BINARY_INV)
plt.imshow(img_tresh, cmap='gray')
```

### Get text

```python
text = pytesseract.image_to_string(img_tresh)
print(text)
```
